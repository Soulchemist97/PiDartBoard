## Py Dartboard Pixel ausgeben

```python
import numpy as np

def Winkel_Pixel(Winkel,Radius=1):
    x = Radius * np.cos(Winkel+np.pi/2)
    if np.isclose(x,0):
        x=0
    y =  Radius * np.sin(Winkel+np.pi/2)
    if np.isclose(y,0):
        y=0
    return x,y
```

```python
import matplotlib.pyplot as plt 

Score_Values = [20,1,18,4,13,6,10,15,2,17,3,19,7,16,8,11,14,9,12,5]

Koordinaten = [] 

for i,score in enumerate(Score_Values):
    angle = i*np.pi/10

    Koordinate = Winkel_Pixel(angle,400)
    Koordinaten.append(Koordinate)
    print(Koordinate)

Koordinaten = np.array(Koordinaten)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(Koordinaten[:,0],Koordinaten[:,1])

# plt.savefig("Koodinaten_Raster.png",transparent = True,dpi=150)
```
